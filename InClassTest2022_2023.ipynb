## Problem 2
The plot below shows the results of a heading discrimination task where a participant was presented linear heading movements at difference angles to the left and right straight ahead(0 degree) with visual alone, vestibular alone and the combined visual-vestibular.

The plot below shows:
* the visual alone psychometric function in red with a point of subjective equality (PSE) of 5 degrees and a just noticable difference (JND) of 7 degrees.
* the vestibular alone psychometric function  in black with a point of subjective equality (PSE) of 0 degrees and a just noticable difference (JND) of 5 degrees.



__Add the visual-vestibular psychometric function in green with a point of subject equality (PSE) of 2.2 degrees and just noticable difference (JND) of 3.8 degrees.__

Hint: W1T4 The Psychometric for Multisensory Integration.ipynb <a href="https://colab.research.google.com/github/john-s-butler-dit/Basic-Introduction-to-Python/blob/master/W1T4%20The%20Psychometric%20for%20Multisensory%20Integration.ipynb" target="_parent"><img src="https://colab.research.google.com/assets/colab-badge.svg" alt=" Open In Colab"/></a>

```python
Heading=np.arange(-20,20,0.1)


## Visual Psychometric Function
PSE_Vis=5
JND_Vis=7

Psychometric_Vis= norm.cdf(Heading,PSE_Vis,JND_Vis)


## Vestibular Psychometric Function
PSE_Vest=0
JND_Vest=5

Psychometric_Vest= norm.cdf(Heading,PSE_Vest,JND_Vest)



##############################################################################
## INSERT: function for vis-vest psychometric function with a PSE of 2 and JND of 4







##############################################################################

fig = plt.figure(figsize=(6,4)) # This setups the size of the figure
plt.plot(Heading,Psychometric_Vest,':',color='black',label='Vestibular')
plt.plot(Heading,Psychometric_Vis,'-',color='red',label='Visual')

##############################################################################
## INSERT: Plot the vis-vest psychometric function with a PSE of 2 and JND of 4
##############################################################################




plt.legend()
plt.xlabel('Heading')
plt.ylabel('Proportion to the Right')
plt.show() # This plots the figure
```

## Problem 4
Given the plots below:

1)	Write a few lines about what is plotted in Figures A, B and C. 

2)	Describe how the data in Figure A is analysed to produce Figures B and C. 

3)	Write a few lines on how the Figures B and Figure C can be used to decode different neuronal information

```python
Samples=1000
Trials=100
Spike_Rate=20
Random_numbers=np.random.uniform(0,1,[Samples,Trials])

Spike_Probability_Threshold=poisson.pmf(1, Spike_Rate/1000)*np.ones([Samples,1])
Spike_Probability_Threshold[300:500]=poisson.pmf(1, 60/1000)


Spikes=np.zeros([Samples,Trials])

Spikes[Random_numbers<Spike_Probability_Threshold]=1


fig = plt.figure(figsize=(14,4)) # This setups the size of the figure
plt.subplot(131)
plt.pcolor(Spikes.transpose(),cmap='gray')
plt.xlabel('Time (ms)')
plt.ylabel('Trials')
plt.title('Figure A')

plt.subplot(132)
plt.hist(Spikes.sum(0))
plt.xlabel('Spikes per Trial')
plt.ylabel('Frequency')
plt.title('Figure B')

plt.subplot(133)
plt.plot(np.convolve(Spikes.sum(1), np.ones(100)/10, mode='valid'))
plt.xlabel('time (ms)')
plt.ylabel('Spike Rate Hz')
plt.title('Figure C')
plt.show()
```
