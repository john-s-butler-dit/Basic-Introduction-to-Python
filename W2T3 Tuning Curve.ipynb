# Tuning Curve, Week 2 Tutorial 3

In this notebook we will illustrate different neuronal tuning curves.

### Libraries

```python
# LIBRARY

import numpy as np # vector manipulation
from scipy.stats import norm # Psychometric Function
from scipy.stats import poisson

# THIS IS FOR PLOTTING
%matplotlib inline
import matplotlib.pyplot as plt # side-stepping mpl backend
import warnings
warnings.filterwarnings("ignore")
```

## Gaussian Tunning Function

To simulate a Gaussian Tunning curve we need to know the Maximum Firing Rate, the Minimum Firing Rate (Baseline), the Preferred direction and the variance. 


```python
Orientation=np.arange(-40,40,1)
Max_Firing=60
Min_Firing=15
Variance=10
Preferred_Direction=0
tunning_curve= norm.pdf(Orientation,Preferred_Direction,Variance)
tunning_curve= tunning_curve/max(tunning_curve)
tunning_curve= (Max_Firing-Min_Firing)*tunning_curve+Min_Firing
fig = plt.figure(figsize=(8,4)) # This setups the size of the figure
plt.plot(Orientation,tunning_curve,'-',color='black')
plt.xlabel('Orientation')
plt.ylabel('Firing Rate (Hz)')
plt.show() # This plots the figure
```

## Sigmoid Tunning Function
Like the Gaussian Tunning curve, to simulate a Sigmoid Tunning curve we need to know the Maximum Firing Rate, the Minimum Firing Rate (Baseline), the Preferred Disparity and the variance. 



```python
Disparity=np.arange(-1,1,0.025)
Max_Firing=40
Min_Firing=0
Variance=0.2
Preferred_Disparity=0
Disparity_tunning_curve= norm.cdf(Disparity,Preferred_Disparity,Variance)
Disparity_tunning_curve= (Max_Firing-Min_Firing)*Disparity_tunning_curve+Min_Firing


fig = plt.figure(figsize=(8,4)) # This setups the size of the figure
plt.plot(Disparity,Disparity_tunning_curve,'-',color='black')
plt.xlabel('Disparity')
plt.ylabel('Firing Rate (Hz)')
plt.show() # This plots the figure
```
